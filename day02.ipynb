```python
import pandas as pd
```

```python
df = pd.read_excel("usd_rate.xlsx")
```

```python
df.head()
```

```python
df.curs.plot()
```

```python
money = df.curs
```

```python
past = 4 * 7 # Возьмем данные за 4 недели в прошлом
future = 7 # Попытаемся предсказать курс на 7 дней в будущее
```

```python
money.describe()
```

```python
start = past
end = len(money) - future
```

```python
print(start, end)
```

```python
new_df = []
for i in range(start, end):
    all_columns = money[(i-past):(i+future)]
    new_df.append(list(all_columns))
```

```python
past_columns = []
for i in range(past):
    past_columns.append(f"past_{i}")
print(past_columns)
```

```python
future_columns = []
for i in range(future):
    future_columns.append(f"future_{i}")
print(future_columns)
```

```python
transformed_df = pd.DataFrame(new_df, columns=(past_columns+future_columns))
```
